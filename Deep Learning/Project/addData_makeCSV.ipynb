```python
import os
from tqdm import tqdm
import pandas as pd
import numpy as np
```

```python
path = "./data/whiteblack/"
file_list = os.listdir(path)
```

```python
len(file_list)
```

```python
file_list
```

```python
img_list = []
label_list = []
```

```python
for name in tqdm(file_list):
    img_path = 'whiteblack/'+name
    if not '0' <= name[0] and name[0] <= '9': continue
    label = int(name[0])
    img_list.append(img_path)
    label_list.append(label)
```

```python
df = pd.DataFrame([ x for x in zip(img_list, label_list)])
df.rename(columns = {0:'img_path',1:'label'},inplace=True)
df.to_csv('./data/whiteblack.csv', index=False)
df
```
